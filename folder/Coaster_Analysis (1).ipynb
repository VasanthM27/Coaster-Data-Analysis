```python
df1 = df[['coaster_name',
    #'Length', 'Speed',
    'Location', 'Status',
    # 'Opening date',
      # 'Type',
    'Manufacturer',
    #'Height restriction', 'Model', 'Height',
     #  'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
      # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
      # 'Track layout', 'Fastrack available', 'Soft opening date.1',
      # 'Closing date', 'Opened', 'Replaced by', 'Website',
      # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
      # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
     'latitude', 'longitude',
     'Type_Main',
       'opening_date_clean',
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    # 'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']]
```

```python
df1.columns
```

```python
df2 = df1.rename(columns={'coaster_name':'Coaster_Name',
                    'year_introduced': 'Year_Introduced',
                    'opening_date_clean': 'Opening_Date',
                    'speed_mph':'Speed_mph',
                    'height_ft': 'Height_ft',
                    'Inversion_clean':'Inversions',
                    'Gforce_clean':'Gforce'})
```

```python
df2.isna().sum()
```

to see which column is duplicated

```python
df2.loc[df2.duplicated(subset=['Coaster_Name'])].head(5)
```

```python
df2.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
final_df = df2.loc[~df2.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
```

Plottings


```python
ax = final_df['Year_Introduced'].value_counts() \
.head(10) \
.plot(kind='bar', title='Top 10 Years Coasters Introduced')

ax.set_xlabel('Year_Introduced')
ax.set_ylabel('Count')
```

```python
final_df['Speed_mph'].plot(kind='hist', bins = 21, title = "coaster speed (mph)")
ax.set_xlabel('Speed (mph)')
```

```python
final_df['Speed_mph'].plot(kind='kde', title = "coaster speed (mph)")
ax.set_xlabel('Speed (mph)')
```

Feature Relationships


```python
final_df.plot(kind='scatter', x='Speed_mph',
              y='Height_ft',
              title='Coaster Speed vs. Height')

plt.show()
```

```python
sns.scatterplot(x='Speed_mph',
                y='Height_ft',
                hue='Year_Introduced',
                data=final_df)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
sns.pairplot(final_df,
             vars=['Year_Introduced', 'Speed_mph',
             'Height_ft','Inversions_clean','Gforce'],
             hue='Type_Main')
plt.show()
```

```python
df_corr = final_df[['Year_Introduced', 'Speed_mph',
             'Height_ft','Inversions_clean','Gforce']].dropna().corr()
```

```python
sns.heatmap(df_corr, annot=True)
```

Asking Questions to our data


```python
final_df.query('Location != "Other"') \
.groupby('Location')['Speed_mph'] \
.agg(['mean','count']) \
.query('count >= 10') \
.sort_values('mean')['mean'] \
.plot(kind='barh', figsize=(12, 5), title= 'Average Coast Speed by Location')

ax.set_xlabel('Average Coaster Speed')
plt.show()
```
